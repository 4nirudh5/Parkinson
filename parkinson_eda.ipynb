## **Univariate Analysis**

```python
for i, predictor in enumerate(df.drop(columns=['status'])):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='status')
```

## **Bivariate Analysis**

```python
g = sns.pairplot(df, diag_kind="kde", markers="+",
                 plot_kws=dict(s=50, edgecolor="b", linewidth=1),
                 diag_kws=dict(shade=True))
```
